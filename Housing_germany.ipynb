# **Visualize Top & Bottom 10 Cities by Cost of Living**

```python
import matplotlib.pyplot as plt

top_cities = germany_df.groupby('city')['cost_of_living'].mean().sort_values(ascending=False).head(10)

top_cities.plot(kind='bar', figsize=(12, 6), color='crimson')
plt.title('Top 10 Most Expensive Cities in Germany')
plt.ylabel('Average Cost of Living (USD)')
plt.xlabel('City')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

# **Bottom 10 Least Expensive Cities**

```python
bottom_cities = germany_df.groupby('city')['cost_of_living'].mean().sort_values(ascending=True).head(10)

bottom_cities.plot(kind='bar', figsize=(12, 6), color='seagreen')
plt.title('Top 10 Least Expensive Cities in Germany')
plt.ylabel('Average Cost of Living (USD)')
plt.xlabel('City')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

### **predictive model to estimate cost of living based on features.**

```python

features = [
    'meal_inexpensive', 'cappuccino', 'bread', 'milk', 'eggs', 'cheese',
    'apt_1bed_city', 'monthly_pass', 'internet'
]

# Target
target = 'cost_of_living'

X = germany_df[features]
y = germany_df[target]
```

```python
# Drop rows with NaNs in X or y
Xy = pd.concat([X, y], axis=1).dropna()

# Split again
X_clean = Xy.drop('cost_of_living', axis=1)
y_clean = Xy['cost_of_living']

# Train-test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_clean, y_clean, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

from sklearn.metrics import mean_squared_error, r2_score
print("MSE:", mean_squared_error(y_test, y_pred))
print("R²:", r2_score(y_test, y_pred))
```

After training and testing the model, we evaluated its performance using two key metrics:

Mean Squared Error (MSE): The model achieved an MSE of 29.82, indicating that the average squared difference between the predicted and actual housing prices is relatively low. This suggests good prediction accuracy.

R² Score (Coefficient of Determination): The R² value was 0.923, meaning that approximately 92.3% of the variation in housing prices can be explained by the input features. This demonstrates that the model fits the data very well and has strong predictive capability.

Overall, the results indicate that the model is effective in capturing the patterns within the data and can reliably estimate housing prices based on the selected features.
